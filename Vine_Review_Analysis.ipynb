```python
import pandas as pd
```

```python
# Read the csv file
vine_df = pd.read_csv("resources/vine_table.csv")
vine_df
```

```python
# Filter DataFrame to locate helpful reviews
threshold_df = vine_df[vine_df['total_votes']>=20]
threshold_df
```

```python
# Filter Dataframe on : helpful votes/total votes >= 0.5
fifty_df = threshold_df[(threshold_df["helpful_votes"] / threshold_df["total_votes"]>= 0.5)]
fifty_df
```

```python
# Retrieve rows where a review was written as part of the Vine Program
paid_df = fifty_df[fifty_df["vine"] == "Y"]
paid_df
```

```python
# Retrieve rows where a review was not part of the Vine Program
free_df = fifty_df[fifty_df["vine"] == "N"]
free_df
```

```python
fivestar_paid_df= paid_df[paid_df['star_rating'] == 5]
fivestar_paid_df
```

```python
fivestar_unpaid_df= free_df[free_df['star_rating']== 5]
fivestar_unpaid_df
```

## Complementary Analysis

```python
# Calculate total Vine reviews in initial dataset
total_vine = vine_df[vine_df['vine']== "Y"].count()
total_vine["vine"]
```

```python
# Calculate total non Vine reviews in initial dataset
total_free = vine_df[vine_df['vine']== "N"].count()
total_free["vine"]
```

```python
# Get count for useful Vine reviews
useful_vine = paid_df['vine'].count()
useful_vine
```

```python
# Get count for useful non Vine reviews
useful_free = free_df['vine'].count()
useful_free
```

```python
# Calculate useful Vine reviews percentage
use_vine = (useful_vine/total_vine)*100
use_vine["vine"]
```

```python
# Calculate useful non Vine reviews percentage
nuse_vine = (useful_free/total_free)*100
nuse_vine["vine"]
```
